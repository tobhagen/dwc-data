```python
df_target = pd.read_csv("https://raw.githubusercontent.com/kfreyburger/sap/main/dwc/v01/D4CaseDataflow/GbiProductAttrTarget.csv", sep = ";", header = 0)
df_target.head()
```

Union aus dem Dataflow simulieren

```python
df_target = df_target.append(df_source)
df_target
```

Jetzt kommt der Code aus dem Dataflow; Ergebnis sind die neuen Sätze, die angefügt werden

```python
max = int(df_target["SurrogatId"].max())
mask1 = df_target["SurrogatId"]==0 
mask2 = ~df_target.duplicated(subset='ProductId', keep=False)
mask = mask1 & mask2
anz_new = len (df_target[mask])
df_target.loc[mask,"SurrogatId"]= range(max+1,max+1+anz_new)
df_target[mask]
```
